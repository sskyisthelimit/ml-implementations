```python
import numpy as np
import numpy.linalg as linal
from sklearn.datasets import load_diabetes
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression as SklearnLinearRegression
import pandas as pd
import matplotlib.pyplot as plt
from LinearRegression import LinearRegression
```

```python
df = pd.read_csv('Admission_Predict.csv')
X, y = df.iloc[:, 1:-1] .to_numpy(), df.iloc[:, -1].to_numpy()
```

```python
X_train = X[:-40]
X_test = X[-40:]


y_train = y[:-40]
y_test = y[-40:]
```

```python
np.shape(X_train), np.shape(y_train)
```

```python
regr = SklearnLinearRegression()
regr.fit(X=X_train, y=y_train)
y_pred = regr.predict(X=X_test)
```

```python
custom_regr = LinearRegression(method="")
custom_regr.fit(X=X_train, y=y_train)
custom_y_pred = custom_regr.predict(X=X_test)
```

```python
plt.figure(figsize=(14, 10))

plt.plot(y_test, label='Actual Values', color='b', marker='o')
plt.plot(custom_y_pred, label='Predicted Values', color='r', linestyle='--', marker='x')

plt.xlabel('Sample Index')
plt.ylabel('Values')
plt.title('Comparison of Actual and Predicted Values')
plt.legend()
plt.grid(True)

plt.show()
```
