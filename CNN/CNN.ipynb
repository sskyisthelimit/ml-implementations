```python
import numpy as np

from keras.datasets import mnist
from keras.utils import to_categorical

from CNN import (ConvolutionalLayer, Reshape, 
                     calc_bin_cross_entropy_grad, 
                     calc_bin_cross_entropy,
                     Sigmoid, Dense)

from CNN import predict, train
```

```python
def preprocess_data(x, y, lim):
    three_i = np.where(y == 3)[0][:lim]
    nine_i = np.where(y == 9)[0][:lim]
    all_indices = np.hstack((three_i, nine_i))
    all_indices = np.random.permutation(all_indices)
    x, y = x[all_indices], y[all_indices]
    x = x.reshape(len(x), 1, 28, 28)
    x = x.astype('float32') / 255
    y = (y == 9).astype(int)
    y = to_categorical(y, 2)
    return x, y
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, y_train = preprocess_data(x_train, y_train, 100)
x_test, y_test = preprocess_data(x_test, y_test, 100)
```

```python
cnn = [
    ConvolutionalLayer((1, 28, 28), 3, 5),
    Sigmoid(),
    Reshape((5, 26, 26), (5*26*26, 1)), 
    Dense(5*26*26, 100),
    Sigmoid(),
    Dense(100, 2),
    Sigmoid()
]

train(
    cnn,
    calc_bin_cross_entropy,
    calc_bin_cross_entropy_grad,
    x_train,
    y_train,
    epochs=20,
    learning_rate=0.1
)

# test
for x, y in zip(x_test, y_test):
    output = predict(cnn, x)
    print(f"pred: {np.argmax(output)}, true: {np.argmax(y)}")
```
