```python
import numpy as  np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from RandomForest import RandomForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
iris = load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)

forest = RandomForest(min_samples=3)
forest.fit(X_train, y_train)

custom_y_pred = forest.predict(X_test)

custom_pred_accuracy = accuracy_score(y_test, custom_y_pred)
print("Custom predicted score accuracy: ", custom_pred_accuracy)

sklearn_forest = RandomForestClassifier()
sklearn_forest.fit(X_train, y_train)
y_pred_sklearn = sklearn_forest.predict(X_test)
sk_pred_accuracy = accuracy_score(y_test, y_pred_sklearn)
print("Sklearn predicted score accuracy: ", sk_pred_accuracy)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

cm_forest = confusion_matrix(y_test, custom_y_pred)
sns.heatmap(cm_forest, annot=True, fmt='d', ax=axes[0], cmap='Blues')
axes[0].set_title('Random Forest Confusion Matrix')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('True')

sk_forest = confusion_matrix(y_test, y_pred_sklearn)
sns.heatmap(sk_forest, annot=True, fmt='d', ax=axes[1], cmap='Blues')
axes[1].set_title('Sklearn Random Forest Confusion Matrix')
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('True')

plt.tight_layout()
plt.show()
```
